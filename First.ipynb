```python
df_nei = pd.read_csv('DN.csv')
```

```python
df_nei.drop(df_nei.columns[[1,3,5]], axis=1,inplace=True)
df_nei=df_nei.drop(0, axis=0)
df_nei.head()
```

```python
df_nei.reset_index(drop=True,inplace=True)
df_nei.head()
```

```python
df_nei.columns = ['Districts','Neighbourhood','Latitude','Longitude']
```

```python
address = 'Delhi, INDIA'

geolocator = Nominatim(user_agent="Jupyter")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of New York City are {}, {}.'.format(latitude, longitude))
```

```python
# create map of Toronto using latitude and longitude values
map_Delhi = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, districts, Neighbourhood in zip(df_nei['Latitude'], df_nei['Longitude'], df_nei['Districts'], df_nei['Neighbourhood']):
    label = '{}, {}'.format(Neighbourhood, districts)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Delhi)  
    
map_Delhi
```
